```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
import re
import os
from PIL import Image
import requests
from io import BytesIO
from joblib import Parallel, delayed

newpath = 'external_data2/' 
if not os.path.exists(newpath):
    os.makedirs(newpath)

print(newpath)

def get_html(url):
    response = requests.get(url)
    return response.text

# from Tomomi 
# https://www.kaggle.com/c/human-protein-atlas-image-classification/discussion/69984
df = pd.DataFrame.from_csv('HPAv18Y.csv', index_col = None)

names = []
urls = []
for name, y in df[['Id','GotYellow']].values: 
    sn = name.split('_')
    last = '_'.join(sn[2:])
    colors = ['red','green','blue']
    if y == 1: colors += ['yellow']
    for c in colors:
        url = 'http://v18.proteinatlas.org/images/' + sn[1] + '/' + last + '_' + c + '.jpg'
        urls.append(url)
        names.append(name + '_' + c + '.jpg')
        
print(len(urls),len(names))
assert len(urls) == len(names)
                
def load_img(url):
    html = get_html(url)
    soup = BeautifulSoup(html, 'lxml')
    links = []
    for a in soup.findAll('a',{'class':'colorbox'},href=True):
        if '_selected' in  a['href']:
            links.append(''.join(('https://www.proteinatlas.org'+a['href']).split('_medium')))
    
    i = 0
    for link in set(links):
        try:
            name = url.split('/')[-2]        
            response = requests.get(link)
            img = Image.open(BytesIO(response.content))
            if np.array(img)[:,:,0].mean()<70:
                img.save(new_path+name+'_'+str(i)+'.png')
                i+=1
                #break
        except:
            pass
```
